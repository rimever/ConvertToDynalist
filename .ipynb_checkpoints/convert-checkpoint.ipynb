```python
stack = []
with open('./data.txt') as f:
    line = f.readline()
    while line:
        line = f.readline()
        if not line:
            for s in stack:
                print(s)
                stack.clear()
        else:
            if len(stack) > 0:
                stack += "   * " + line + "\n"
            else:
                stack += "* " + line + "\n"


for s in stack:
    print(s)
```
